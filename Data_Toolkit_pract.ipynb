**3. Explain the following terms:**

The Difference Between np.array, np.asarray, and np.asanyarray
These functions are used to convert input data (like lists, tuples, or other arrays) into NumPy arrays, but they have subtle differences in behavior.

 np.array- creates a new array object from the input data. It copies the data by default unless specified otherwise.
 Always returns a new array, even if the input is already an array.
 By default, it creates a copy of the input data.

 np.asarray converts the input to an array but does not make a copy if the input is already an array of the correct type.
 Avoids copying the input data when possible. Useful for functions that should accept arrays without duplicating data.
 Only creates a copy if necessary (i.e., if the input is not already an array or if the input array has a different dtype).

 np.asanyarray is similar to np.asarray but it passes through subclasses of np.ndarray (like matrix, masked_array, etc.).
 Converts input to an array but preserves any subclasses of ndarray.
 Avoids copying the input data when possible and preserves subclasses.

```python
#np.array
lst = [1, 2, 3]
arr = np.array(lst)
print(arr)
```

```python
#np.asarray
lst = [1, 2, 3]
arr = np.asarray(lst)
print(arr)

existing_arr = np.array([4, 5, 6])
arr2 = np.asarray(existing_arr)
print(arr2 is existing_arr)
```

```python
#np.asanyarray
lst = [1, 2, 3]
arr = np.asanyarray(lst)
print(arr)

existing_arr = np.array([4, 5, 6])
arr2 = np.asanyarray(existing_arr)
print(arr2 is existing_arr)

masked_arr = np.ma.array([7, 8, 9], mask=[0, 1, 0])
arr3 = np.asanyarray(masked_arr)
print(type(arr3))
```

The Difference Between Deep Copy and Shallow Copy-
These concepts are essential for understanding how data structures are copied in programming, especially mutable ones like lists and dictionaries.

Shallow Copy- A shallow copy creates a new object, but inserts references into it to the objects found in the original. It means that nested objects (like lists within lists) are not copied but only their references are.
Changes to mutable nested objects affect both the original and the copied object.

```python
import copy

original_list = [1, 2, [3, 4]]
shallow_copied_list = copy.copy(original_list)

original_list[2][0] = 'changed'
print(original_list)
print(shallow_copied_list)
```

Deep Copy- A deep copy creates a new object and recursively inserts copies of the objects found in the original. This results in completely independent objects. Changes to the original object or its nested objects do not affect the deep copy and vice versa.

```python
import copy

original_list = [1, 2, [3, 4]]
deep_copied_list = copy.deepcopy(original_list)

original_list[2][0] = 'changed'
print(original_list)
print(deep_copied_list)
```

Shallow Copy: Copies the object but not the nested objects (only references).

Deep Copy: Copies the object and all objects nested within it, creating independent copies.

**6. Create a 3D NumPy array of shape (3,3,3) containing random integers between 1 and 10. Perform the following operations:**

 a) Find the indices of the maximum values along each depth level (third axis).

 b) Perform element-wise multiplication of between both array.

Sol: Generate the 3D random array: Use numpy.random.randint() to generate random integers.

Find indices of the maximum values along each depth level: Use numpy.argmax() with the appropriate axis.

Perform element-wise multiplication: Multiply the original array by itself or another array of the same shape.

**14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:**

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a time-series dataset with 'Date', 'Temperature', and 'Humidity'
date_range = pd.date_range(start='2024-01-01', periods=30, freq='D')  # 30 days starting from Jan 1, 2024
temperature = np.random.uniform(15, 30, size=len(date_range))  # Random temperatures between 15 and 30
humidity = np.random.uniform(40, 80, size=len(date_range))  # Random humidity between 40% and 80%

# Create a DataFrame
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# a) Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# b) Label the x-axis as 'Date' (already done above)

# c) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

**15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:**

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Creating a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plotting a histogram of the data with 30 bins
fig, ax = plt.subplots(figsize=(10, 6))
count, bins, _ = ax.hist(data, bins=30, density=True, alpha=0.6, color='b', label='Histogram')

# Overlaying a line plot representing the normal distribution's PDF
mu, std = norm.fit(data)  # Fitting a normal distribution to the data
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
ax.plot(x, p, 'k', linewidth=2, label='PDF')

# Labeling the x-axis and y-axis
ax.set_xlabel('Value')
ax.set_ylabel('Frequency/Probability')

# Setting the title of the plot
plt.title('Histogram with PDF Overlay')

# Adding a legend
plt.legend()

# Showing the plot
plt.show()
```

**17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating two random arrays
np.random.seed(42)
x = np.random.randn(100)
y = np.random.randn(100)

# Determining the quadrant based on the position relative to the origin
quadrant = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        quadrant.append('Q1')
    elif x[i] < 0 and y[i] >= 0:
        quadrant.append('Q2')
    elif x[i] < 0 and y[i] < 0:
        quadrant.append('Q3')
    else:
        quadrant.append('Q4')

# Creating a Seaborn scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, hue=quadrant, palette='Set1', s=100)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Adding a legend
plt.legend(title='Quadrant')

# Showing the plot
plt.show()
```

**18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Create data for the sine wave
x = np.linspace(0, 4 * np.pi, 1000) #By extending the range of x from 0 to 4π, you are capturing two full cycles of the sine wave.
y = np.sin(x)
# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)
# Grid lines are automatically added, but you can customize them
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"
p.xgrid.grid_line_dash = [6, 4]
p.ygrid.grid_line_dash = [6, 4]
# Display the plot
show(p)
```

**19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Generate random categories and corresponding values
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame for easier handling
data = pd.DataFrame({'categories': categories, 'values': values})

source = ColumnDataSource(data)

# Create a new figure with title and axis labels
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value',
           toolbar_location=None, tools="")

# Add a bar renderer with a color map based on the values
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white',
       fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

hover = HoverTool()
hover.tooltips = [
    ("Category", "@categories"),
    ("Value", "@values")
]
p.add_tools(hover)
# Remove gridlines
p.xgrid.grid_line_color = None
p.y_range.start = 0  # Start the y-axis from 0

# Show the plot
show(p)
```

**20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.**


```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)
y = np.random.randn(100).cumsum()

# Create the figure
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout with axis labels and title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

# Show the plot
fig.show()
```
